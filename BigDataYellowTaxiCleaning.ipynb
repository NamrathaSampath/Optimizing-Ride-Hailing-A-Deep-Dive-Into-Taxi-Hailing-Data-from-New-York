```python
import pandas as pd
import numpy as np
```

```python
import pyarrow.parquet as pq
trips = pq.read_table('/yellow_tripdata_2022-01.parquet')
trips = trips.to_pandas()
```

```python
trips.head()
```

```python
print(trips.dtypes)
```

```python
trips.rename(columns={
    'VendorID': 'vendor_id',
    'tpep_pickup_datetime': 'pickup_datetime',
    'tpep_dropoff_datetime': 'dropoff_datetime',
    'extra': 'extra_charges',
    'mta_tax': 'tax',
    'tolls_amount': 'toll_amount',
    'total_amount': 'total_fare'
}, inplace=True)
```

```python
trips = trips.drop(['passenger_count', 'RatecodeID', 'store_and_fwd_flag'], axis=1)
```

```python
trips['extra_charges'] = trips['extra_charges'] + trips['improvement_surcharge']
```

```python
trips.head()
```

```python
trips = trips.drop(['improvement_surcharge'], axis=1)
```

```python
trips.head()
```

```python
trips['driver_pay'] = trips['total_fare']
```

```python
trips['pickup_datetime'] = pd.to_datetime(trips['pickup_datetime'])
trips['dropoff_datetime'] = pd.to_datetime(trips['dropoff_datetime'])

# Calculate 'trip_duration' in seconds
trips['trip_duration'] = (trips['dropoff_datetime'] - trips['pickup_datetime']).dt.total_seconds()

# Display the DataFrame with the new 'trip_duration' column
print(trips[['pickup_datetime', 'dropoff_datetime', 'trip_duration']])
```

```python
# Convert 'pickup_datetime' to datetime objects
trips['pickup_datetime'] = pd.to_datetime(trips['pickup_datetime'])

# Create new columns equal to 'pickup_datetime'
trips['request_datetime'] = trips['pickup_datetime']
trips['on_scene_datetime'] = trips['pickup_datetime']
```

```python
trips['vendor_commission'] = 0
```

```python
trips.head()
```

```python
trips = trips.reindex(columns=['vendor_id','pickup_datetime','dropoff_datetime','trip_distance','PULocationID','DOLocationID','payment_type','fare_amount','extra_charges','tax','tip_amount','toll_amount','total_fare','congestion_surcharge','airport_fee','driver_pay','trip_duration','request_datetime','on_scene_datetime','vendor_commission'])
```
